```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import math
import re
%matplotlib inline
import chart_studio.plotly as py
import plotly.offline as py
from plotly.offline import iplot
from plotly import tools
import plotly.tools as tls
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import cufflinks as cf
cf.go_offline()
```

```python
calendar = pd.read_csv('calendar.csv')
```

```python
calendar.head()
```

```python
calendar.price = calendar.price.apply(lambda i: int(re.findall(r'\d+', i)[0]))
calendar.adjusted_price = calendar.adjusted_price.apply(lambda i: int(re.findall(r'\d+', i)[0]))
```

```python
calendar[['minimum_nights']] = calendar[['minimum_nights']].astype(int)
calendar[['maximum_nights']] = calendar[['maximum_nights']].astype(int)
```

```python
calendar.info()
```

```python
calendar = calendar[(calendar['date'] <= '2021-04-30')]
calendar
```

```python
#Average price per day
average_price = pd.DataFrame(calendar.groupby(['date']).mean()['price']).reset_index()
average_price
```

```python
average_price.price.mean()
```

```python
trace1 = go.Scatter(x=average_price.date, y=average_price.price, line = dict(color = '#7F7F7F'))

data = [trace1]

layout = {
    'title': 'Average Price Per Day',
    'xaxis': {'title': 'Day'},
    'yaxis': {'title': 'Average Price $'},
               
    'shapes': [
        {
            'type': 'line',
            'x0': '2020-04-28',
            'y0': 166,
            'x1': '2021-04-30',
            'y1': 166,
            'line': {
                'color': 'black',
                'width': 1,
                'dash': 'dashdot',
            }
        }
    ]
}

layout.update(dict(
    annotations=[go.Annotation(text="Average Price $166", x='2020-04-28', y=166)]),
             xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=1,
                    label='YTD',
                    step='year',
                    stepmode='todate'),
                dict(count=1,
                    label='1y',
                    step='year',
                    stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
             )
              

py.iplot({'data': data, 'layout': layout}, filename='Daily Average Price')
```

```python
#Average price per month
calendar['year'], calendar['month'], calendar['day'] = calendar['date'].str.split('-',2).str
calendar
```

```python
calendar['month'] = calendar['year'].map(str) + "-" + calendar['month'].map(str)
calendar.date = pd.to_datetime(calendar['month'])
price_month = calendar.groupby('month').describe()['price'].reset_index()
price_month = price_month.drop(['count', 'std', '25%', '50%', '75%'], axis=1)
price_month
```

```python
trace_mean = go.Scatter(
    x=price_month.month,
    y=price_month['mean'],
    name = "Average Price Per month",
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)

data = [trace_mean]

layout = {
    'title': 'Monthly Price Trend',
    'xaxis': {'title': 'Month'},
    'yaxis': {'title': 'Average Price ($)'},
               
    'shapes': [
        {
            'type': 'line',
            'x0': '2020-04',
            'y0': 166,
            'x1': '2021-04',
            'y1': 166,
            'line': {
                'color': 'black',
                'width': 1,
                'dash': 'dashdot',
            }
        }
    ]
}

layout.update(dict(annotations=[go.Annotation(text="Average Price $166", x='2020-04', y=166)]))
        
fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Monthly Price")
```
